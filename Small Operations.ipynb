```python
#input a tuple of 5 integers. Pass it to function count() all the even elements and returns the ans
def count(num):
    s=[]
    for x in num:
        if(x%2==0):
            s.append(x)
    return tuple(s)

l1 =[]
for i in range(0, 5): 
    ele = int(input()) 
    l1.append(ele)
t1=tuple(l1)
count(t1)
```

```python
#input a book name. Pass it to function count() which counts how many upper case and lower case characters are present in the string.
def count(string): 
  
    upper = 0
    lower = 0
  
    for i in range(len(string)): 
          
        # For lower letters 
        if (ord(string[i]) >= 97 and
            ord(string[i]) <= 122): 
            lower += 1
  
        # For upper letters 
        elif (ord(string[i]) >= 65 and
              ord(string[i]) <= 90): 
            upper += 1
  
    print('Lower case characters = %s' %lower, 
          '\nUpper case characters = %s' %upper) 

name = input("Enter book name: ")
count(name)
```

```python
#input a string and a word and chick wheteher if the word exists in the string through the fuction check().
def check(string,word):
    l1=[]
    l1 = list(string.split(" ")) 
    for x in l1:
        if word == x:
            print("The word exists, it's: ",x)
string=input("Enter String: ")
word=input("Enter the word you wanna check: ")

check(string,word)
```

# Stacks

```python
#Implement Stacks using list, where stack contains names of students.
stack = []
n = int(input("Enter the no of students: "))
#push
for i in range(n): 
        ele =input("Enter Name of the students: ") 
        stack.append(ele) 

print('Initial stack')
print(stack)

#pop
print('\nElements poped from stack:')
for i in range(n):
    print(stack.pop(0))
 
print('\nStack after elements are poped:')
print(stack)
```
